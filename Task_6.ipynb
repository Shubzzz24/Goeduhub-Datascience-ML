**1)Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a car’s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.**

         Days=[1,2,3,4,5,6,7,8]

         Speed=[60,62,61,58,56,57,46,63]​​​​​

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Days=[1,2,3,4,5,6,7,8]
Speed=[60,62,61,58,56,57,46,63]
plt.figure(figsize=(12,6))
plt.plot(Days,Speed,linewidth=4)
plt.title("Car Speed Measurement")
plt.xlabel("Days")
plt.ylabel("Speed")
plt.show()
```

**2)Now to above car data apply some string formats  like line style example green dotted line, marker shape like +, change markersize, markerface color etc.**

```python
plt.figure(figsize=(12,6))
plt.plot(Days,Speed,linewidth=4,linestyle='dotted',color='green',marker='+',ms=15,mec='r')
plt.show()
```

**4) Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.**

```python
x = np.linspace(0,6,100)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(12,6))
plt.plot(x,y,':b',linewidth=3)
plt.plot(x,z,'--r',linewidth=3)
plt.xlabel("Angle in radian")
plt.ylabel("Magnitude")
plt.legend(['Sine','Cosine'])
plt.show()
```

**5)Plot Simple bar chart showing popularity of Programming Languages.**

Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

Popularity = [56, 39, 34, 34, 29]

**Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function**

```python
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]
plt.figure(figsize=(12,6))
x_axis = np.arange(len(Languages))

plt.bar(x_axis-0.2 ,Popularity,0.4,label="Popularity")
plt.bar(x_axis+0.2 ,Security,0.4,label="Security")

plt.xticks(x_axis,Languages)
plt.xlabel("Languages")
plt.ylabel("Range")
plt.title("Languages/Popularity,Security")
plt.legend()
plt.show()
```

```python
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]

plt.figure(figsize=(12,6))
x_axis = np.arange(len(Languages))

plt.barh(x_axis-0.2 ,Popularity,0.4,label="Popularity")
plt.barh(x_axis+0.2 ,Security,0.4,label="Security")

plt.yticks(x_axis,Languages)
plt.xlabel("Range")
plt.ylabel("Languages")
plt.title("Languages/Popularity,Security")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import pandas as pd

dataframe = pd.read_csv('titanic')
```

```python
dataframe.head()
```

```python
plt.figure(figsize=(9,9))
dataframe.sex.value_counts().plot.pie(autopct='%.0f%%',labels=['Male','Female'],shadow=True,startangle=180)
plt.title('classification')
plt.show()
```

```python
df_male = dataframe[dataframe['sex'] == 'male']
df_male.sex.value_counts()                
```

```python
df_female = dataframe[dataframe['sex'] == 'female']
df_female.sex.value_counts()
```

```python
plt.figure(figsize=(9,9))
plt.scatter(df_male.fare,df_male.age,marker='+')
plt.scatter(df_female.fare,df_female.age,marker='.')
plt.title('Scatter plot')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.show()
```
